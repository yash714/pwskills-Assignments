```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, StandardScaler

# Load the dataset
df = pd.read_csv('dataset.csv')
```

```python
"""
Q2. Split the dataset into a training set (70%) and a test set (30%).
"""
```

```python
from sklearn.model_selection import train_test_split

X = df.drop('target', axis=1)
y = df['target']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X,y
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

```

```python
"""
Q3. Train a random forest classifier on the training set using 100 trees and a maximum depth of 10 for each
tree. Use the default values for other hyperparameters.
"""
```

```python
from sklearn.ensemble import RandomForestClassifier

# Create a random forest classifier
rf = RandomForestClassifier(n_estimators=100, max_depth=10, random_state=42)

# Fit the classifier on the training set
rf.fit(X_train, y_train)
```

```python
"""
Q4. Evaluate the performance of the model on the test set using accuracy, precision, recall, and F1 score.
"""
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


# Make predictions on the test set
y_pred = rf.predict(X_test)

# Calculate the accuracy, precision, recall, and F1 score
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

# Print the evaluation metrics
print("Accuracy: {:.2f}".format(accuracy))
print("Precision: {:.2f}".format(precision))
print("Recall: {:.2f}".format(recall))
print("F1 score: {:.2f}".format(f1))
```

```python
"""
Q5. Use the feature importance scores to identify the top 5 most important features in predicting heart
disease risk. Visualise the feature importances using a bar chart.
"""
```

```python
import matplotlib.pyplot as plt
import numpy as np
# Extract the feature importances
importances = rf.feature_importances_

# Sort the features by importance
indices = np.argsort(importances)[::-1]

# Get the top 5 most important features
top_features = X_train.columns[indices][:5]

# Print the top 5 most important features
print("Top 5 most important features:")
for feature in top_features:
    print("- {}".format(feature))

# Visualize the feature importances using a bar chart
plt.bar(range(X_train.shape[1]), importances[indices])
plt.xticks(range(X_train.shape[1]), X_train.columns[indices], rotation=90)
plt.xlabel("Features")
plt.ylabel("Importance score")
plt.title("Feature importances")
plt.show()
```

```python
"""
Q6. Tune the hyperparameters of the random forest classifier using grid search or random search. Try
different values of the number of trees, maximum depth, minimum samples split, and minimum samples
leaf. Use 5-fold cross-validation to evaluate the performance of each set of hyperparameters.
"""
```

```python
from sklearn.model_selection import GridSearchCV

# Define the parameter grid to search over
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [5, 10, 15],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Create a random forest classifier
rf = RandomForestClassifier(random_state=42)

# Use 5-fold cross-validation to evaluate each combination of hyperparameters
grid_search = GridSearchCV(rf, param_grid, cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)

# Print the best hyperparameters and the corresponding mean cross-validation score
print("Best hyperparameters: ", grid_search.best_params_)
print("Cross-validation score: {:.2f}".format(grid_search.best_score_))
```

```python
"""
Q7. Report the best set of hyperparameters found by the search and the corresponding performance
metrics. Compare the performance of the tuned model with the default model.
"""
```

```python
# Create a random forest classifier with the best hyperparameters found by the grid search
rf_tuned = RandomForestClassifier(n_estimators=150, max_depth=5, min_samples_split=5, min_samples_leaf=1, random_state=42)

# Train the model on the training set
rf_tuned.fit(X_train, y_train)

# Evaluate the performance of the tuned model on the test set
y_pred_tuned = rf_tuned.predict(X_test)
acc_tuned = accuracy_score(y_test, y_pred_tuned)
prec_tuned = precision_score(y_test, y_pred_tuned)
rec_tuned = recall_score(y_test, y_pred_tuned)
f1_tuned = f1_score(y_test, y_pred_tuned)

# Print the performance metrics for the tuned model
print("Tuned model performance:")
print("- Accuracy: {:.2f}".format(acc_tuned))
print("- Precision: {:.2f}".format(prec_tuned))
print("- Recall: {:.2f}".format(rec_tuned))
print("- F1 score: {:.2f}".format(f1_tuned))

# Evaluate the performance of the default model on the test set
y_pred_default = rf.predict(X_test)
acc_default = accuracy_score(y_test, y_pred_default)
prec_default = precision_score(y_test, y_pred_default)
rec_default = recall_score(y_test, y_pred_default)
f1_default = f1_score(y_test, y_pred_default)

# Print the performance metrics for the default model
print("Default model performance:")
print("- Accuracy: {:.2f}".format(acc_default))
print("- Precision: {:.2f}".format(prec_default))
print("- Recall: {:.2f}".format(rec_default))
print("- F1 score: {:.2f}".format(f1_default))
```

```python
"""
Q8. Interpret the model by analysing the decision boundaries of the random forest classifier. Plot the
decision boundaries on a scatter plot of two of the most important features. Discuss the insights and
limitations of the model for predicting heart disease risk.
"""
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions

# Get the indices of the two most important features (thalach and cp)
feature_indices = np.argsort(rf_tuned.feature_importances_)[::-1][:2]

# Select the corresponding columns from the dataset
X_plot = X_test[:, feature_indices]

# Create a meshgrid of points to visualize the decision boundaries
xx, yy = np.meshgrid(np.linspace(X_plot[:,0].min()-1, X_plot[:,0].max()+1, 100),
                     np.linspace(X_plot[:,1].min()-1, X_plot[:,1].max()+1, 100))

# Create a new classifier with only two features to plot the decision boundaries
rf_plot = RandomForestClassifier(n_estimators=150, max_depth=5, min_samples_split=5, min_samples_leaf=1, random_state=42)
rf_plot.fit(X_plot, y_test)

# Plot the decision boundaries
fig, ax = plt.subplots(figsize=(10, 8))
plot_decision_regions(X_plot, y_test, clf=rf_plot, legend=2, ax=ax,
                      filler_feature_values={2: 0}, highlight=False, alpha=0.7)

# Add axis labels and title
plt.xlabel("Maximum heart rate achieved")
plt.ylabel("Chest pain type")
plt.title("Decision boundaries of random forest classifier")

# Show the plot
plt.show()
```
